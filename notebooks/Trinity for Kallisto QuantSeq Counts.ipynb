## Run kallisto on all trimmed quantseq files

```python
! ls ../../data/tagseq_trim/*.trim | xargs basename -s _L006_R1_001.fastq.trim
```

```python
%%bash
find ../../data/tagseq_trim/*.trim \
| xargs basename -s .fastq.trim | xargs -I{} \
/Applications/bioinformatics/trinityrnaseq-Trinity-v2.8.3/util/align_and_estimate_abundance.pl \
--transcripts ../../results/trinity/20181029-olurida-gonad-assembly_clust.fasta \
--prep_reference \
--seqType fq \
--single ../../data/tagseq_trim/{}.fastq.trim \
--output_dir {}/ \
--est_method kallisto
```

The above script uses find and xargs basename -s to search for all .trim files, extract the file name that's not redundant (at least the end of the file name, some redundancy remains in the beginning of the names), then do the pseudoalignment using the perl script, and write the output files to directories that are named using the file name.
